```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heart_disease_data = pd.read_csv('heart-disease.csv')
heart_disease_data
```

```python
heart_disease_data['age'].plot.hist();
```

```python
heart_disease_data.plot.hist(figsize=(10,20), subplots=True);
```

```python
over_20 = heart_disease_data[heart_disease_data['age']>20]
over_20.head()
```

```python
#pyplot method
over_20.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
```

```python
# OO method for advance
fig, ax = plt.subplots(figsize=(10,6))
over_20.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_20['age'],
                    y=over_20['chol'],
                    c=over_20['target'])
ax.set(title='heart disease and cholestrol level',
            xlabel='age',
            ylabel='cholestrol')
ax.legend(*scatter.legend_elements(), title='target')

ax.axhline(over_20['chol'].mean(), linestyle='-.')
```
