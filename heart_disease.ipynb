```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heart_disease_data = pd.read_csv('heart-disease.csv')
heart_disease_data
```

```python
heart_disease_data['age'].plot.hist();
```

```python
heart_disease_data.plot.hist(figsize=(10,20), subplots=True);
```

```python
over_20 = heart_disease_data[heart_disease_data['age']>20]
over_20.head()
```

```python
#pyplot method
over_20.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
```

```python
# OO method for advance
fig, ax = plt.subplots(figsize=(10,6))
over_20.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_20['age'],
                    y=over_20['chol'],
                    c=over_20['target'])
ax.set(title='heart disease and cholestrol level',
            xlabel='age',
            ylabel='cholestrol')
ax.legend(*scatter.legend_elements(), title='target')

ax.axhline(over_20['chol'].mean(), linestyle='-.')
```

```python
#make a multiple subplot
#ax0 chol and age
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)
ax0.scatter(x=over_20['age'],
           y=over_20['chol'],
           c=over_20['target'])

ax0.set(title='Heart Disease and Cholesterol Levels',
       ylabel='Cholesterol')

ax0.legend(*scatter.legend_elements(), title='Target')

ax0.axhline(over_20['chol'].mean(), linestyle='--')

#ax1
ax1.scatter(x=over_20['age'],
           y=over_20['thalach'],
           c=over_20['target'])
ax1.set(title='Heart Disease and Max Heart Rate Levels',
       xlabel='Age',
       ylabel='Max Heart Rate Levels')

ax1.legend(*scatter.legend_elements(), title='Target')

ax1.axhline(over_20['thalach'].mean(), linestyle='--')

fig.suptitle('Heart Disease Analysis', 
             fontsize=20,
            fontweight='bold')
```
